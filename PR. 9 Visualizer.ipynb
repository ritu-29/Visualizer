```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class SalesDataAnalyzer:
    def __init__(self):
        self.data = None
        self.numpy_array=np.array([])
        self.last_plot = None 
    
    def load_data(self, file_path):
        try:
            self.data = pd.read_csv(file_path)
            self.numpy_array = self.data.to_numpy()
            print("Dataset loaded successfully!")
        except FileNotFoundError:
            print("File not found!")
    
    def explore_data(self):
        if self.data is None:
            print("No dataset loaded!")
            return
        while True:
            print("\n== Explore Data ==")
            print("1. First 5 rows")
            print("2. Last 5 rows")
            print("3. Columns")
            print("4. Data types")
            print("5. Info")
            print("6.Go back")
    
            ch = int(input("Enter choice: "))
            if ch == 1:
                print(self.data.head())
            elif ch == 2:
                print(self.data.tail())
            elif ch == 3:
                print(self.data.columns)
            elif ch == 4:
                print(self.data.dtypes)
            elif ch == 5:
                self.data.info()
            elif ch == 6:
                break
            else:
                print("invalid")

    def convert_numpy(self):
        if self.data.empty:
            print("Load data first.")
            return
    
        try:
            print(self.data.columns)
            cols = input("Enter column names to convert to NumPy (comma-separated, or leave blank for all): ").strip()
            if cols:
                cols = [c.strip() for c in cols.split(',')]
                self.numpy_array = self.data[cols].values
            else:
                self.numpy_array = self.data.values
            print("\nData successfully converted into a NumPy array.") 
            print(self.numpy_array)
            
            r = input("Enter row index: ").strip()
            c = input("Enter column index: ").strip()
            if r.isdigit() and c.isdigit():
                r = int(r)
                c = int(c)
                print("Value:", self.numpy_array[r, c])
            else:
                print("Indexing Skipping")
                        
        except:
            ("Not Found")

    def combine_data(self):
        file_path = input("Enter CSV path of dataset to combine: ").strip()
        try:
            other_data = pd.read_csv(file_path)
            self.data = pd.concat([self.data, other_data], ignore_index=True)
            print(f"Datasets combined. Total rows: {len(self.data)}")
        except Exception as e:
            print(f"Error: {e}")

    def remove_duplicates(self):
        if self.data.empty:
            print("Load data first.")
            return
    
        before = len(self.data)
        self.data.drop_duplicates(inplace=True)
        after = len(self.data)
    
        print(f"Duplicates removed: {before - after}")
        print(f"Total rows now: {after}")

            
    def search_sort_filter(self):
        if self.data.empty:
            print("Load data first.")
            return
        while True:
            print("1.Search")
            print("2.Sort")
            print("3.filter")
            print("4.back")
            ch = int(input("enter choice"))
            print(self.data.columns)
            if ch == 1:
                col = input("Column to search: ")
                val = input("Value to search: ")
  
                results = self.data[self.data[col].astype(str).str.contains(val, case=False, na=False)]
                
                print(len(results))
                print(results.head(10)) 

            elif ch ==2:
                sort_col = input("\nEnter column name to sort by: ").strip()
                if sort_col in self.data.columns:
                    sorted_df = self.data.sort_values(by=sort_col, ascending=False)
                    print(f"\nTop 5 rows sorted by '{sort_col}' (descending):")
                    print(sorted_df.head())
                else:
                    print(f"Column '{sort_col}' not found.")
            elif ch==3:
                    filter_col = input("Enter numeric column to filter: ").strip()
                    
                    if filter_col in self.data.columns:
                        filter_val = float(input(f"Keep rows where '{filter_col}' >= value: "))
                        filtered_df = self.data[self.data[filter_col] >= filter_val]
                        print(f"\nFiltered data where '{filter_col}' >= {filter_val}:")
                        print(filtered_df.head(10)) 
                    else:
                        print("not found.")
            elif ch==4:
                break

    def groupby_and_transform(self):
        print(self.data.columns)
        if self.data is None or self.data.empty:
            print("Load data first.")
            return
    
        while True:
            print("\n=== GroupBy & Transform Menu ===")
            print("1. GroupBy + Aggregate (sum / mean)")
            print("2. Transform: Percentage of Group Total")
            print("3. Back")
    
            ch = int(input("Enter choice: "))
            if ch == 1:
                group_col = input("Enter column to group by (e.g., Region): ").strip()
                num_col = input("Enter numeric column (e.g., Sales): ").strip()
                agg = input("Aggregation (sum / mean / count): ").strip()
            
                if group_col in self.data.columns and num_col in self.data.columns:
                    result = self.data.groupby(group_col)[num_col].agg(agg)
                    print("Grouped Result:")
                    print(result)
                else:
                    print("Invalid column name.")
            
            elif ch == 2:
                group_col = input("Group by column (e.g., Region): ").strip()
                num_col = input("Numeric column (e.g., Sales): ").strip()
            
                if group_col in self.data.columns and num_col in self.data.columns:
                    self.data["percent_of_group"] = (
                        self.data[num_col] /
                        self.data.groupby(group_col)[num_col].transform("sum")
                    ) * 100
                    print(self.data[[group_col, num_col, "percent_of_group"]].head())
                else:
                    print("Invalid columns.")
            
            elif ch == 3:
                break


    def handle_missing_data(self):
        if self.data is None:
            print("No dataset loaded!")
            return
        while True:
            print("\nHandle Missing Data ")
            print("1. Show missing rows")
            print("2. Fill with mean")
            print("3. Drop missing rows")
            print("4.Go back")
    
            ch = input("Enter choice: ")
    
            if ch == "1":
                print(self.data.isnull().sum())
            elif ch == "2":
                self.data.fillna(self.data.mean(numeric_only=True), inplace=True)
                print("Filled with mean")
            elif ch == "3":
                self.data.dropna(inplace=True)
                print("Dropped missing rows")
            elif ch == "4":
                break

    def generate_descriptive_stats(self):
        if self.data is None:
            print(" No data loaded!")
            return
        while True:
                print("\n--- Statistics Menu ---")
                print("1. Statistical Analysis")
                print("2. Pivot Table")
                print("3. Back to Main Menu")
                ch = int(input("Enter Choice:"))
                if ch == 1:
                    print("Descriptive Statistics (Numeric):")
                    print(self.data.describe())
                    print("Descriptive Statistics (Categorical):")
                    print(self.data.describe(include='object'))
                elif ch == 2:  
                    if self.data.empty:
                        print("Load data first.")
                        return
                    print(self.data.columns)
                    index_cols = input("Enter index columns (comma-separated): ").strip().split(',')
                    value_cols = input("Enter value columns (comma-separated): ").strip().split(',')
                    agg_func = input("Enter aggregation function (sum, mean, count): ").strip()
            
                    index_cols = [c.strip() for c in index_cols]
                    value_cols = [c.strip() for c in value_cols]
                
                    pivot = pd.pivot_table(self.data, index=index_cols, values=value_cols, aggfunc=agg_func)
                    print("\nPivot Table:")
                    print(pivot)

                elif ch == 3:
                    break
                
    def visualize_data(self):
        print(self.data.columns)
        if self.data.empty:
            print("Load data first.")
            return
        sns.set_theme(style="darkgrid")
        while True:
            print("\n====Visualization Menu ====")
            print("1. Bar Plot")
            print("2. Box Plot")
            print("3. Scatter Plot")
            print("4. Histogram ")
            print("5. Heatmap")
            print("6. Pie Chart")
            print("7. Stack Plot")
            print("8. Go Back")
            ch = int(input("Enter Choice:"))
            if ch == 8: 
                break
            try:
                self.last_plot = plt.figure(figsize=(6,4))
                print(self.data.columns)
                if ch == 1:
                    x = input("Enter x-axis column:(e.g., Region) ")
                    y = input("Enter y-axis column:(e.g., Sales) ")
                    a = sns.barplot(x=x, y=y, data=self.data)
                    for i in a.containers:
                        a.bar_label(i,fmt="%.1f")

                elif ch==2:
                    x = input("Enter x-axis column:(e.g., Category) ")
                    y = input("Enter y-axis column:(e.g., Sales) ")
                    a = sns.boxplot(x=x, y=y, data=self.data)

                elif ch==3:
                    x = input("Enter x-axis column name: ")
                    y = input("Enter y-axis column name: ")
                    sns.scatterplot(x=x, y=y, data=self.data)
                    plt.title(f"{y} vs {x}")
                    plt.xlabel(x)
                    plt.ylabel(y)
                    
                elif ch==4:
                    col = input("Enter numeric column for histogram (e.g., Sales): ").strip()
                    bins = int(input("Enter number of bins (e.g., 30): "))
                    sns.histplot(self.data[col], bins=bins, kde=True, color='purple')
                    plt.title(f'Distribution of {col}')
                
                elif ch == 5:
                    num_data = self.data.select_dtypes(include='number')
                    corr = num_data.corr()
                    fig, ax = plt.subplots(figsize=(8,6))
                    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
            
                elif ch==6:
                    col = input("Enter coloumn for pie chart: ")
                    cnt = self.data[col].value_counts().head(10)  
                    plt.pie(cnt.values,labels=cnt.index, autopct="%1.1f%%")
                elif ch==7:
                    cols = self.data.select_dtypes(include='number').columns[:3]
                    plt.stackplot(self.data.index, self.data[cols].T)
                else:
                    print("Invalid")
                    return
                
                plt.title("Visualization")
                self.last_plot = plt.gcf()
                plt.show()
                print("Plot displayed successfully!")
            except:
                ("Invalid")
        
    def save_visualization(self):
        if self.last_plot is None:
            print("No visualization to save!")
            return

        folder = r"C:\Users\Ritu\Desktop\DATA_ANAYLIST"
        name = input("Enter file name (e.g., scatter_plot.png): ")
        full_path = f"{folder}\\{name}"
        
        self.last_plot.savefig(full_path)
        print(f"Visualization saved as {full_path} successfully!")

    def aggregate_functions(self):
        print(self.data.columns)
        if self.data.empty:
            print("Load data first.")
            return
    
        col = input("Enter numeric column to aggregate: ").strip()
        if col in self.data.columns and pd.api.types.is_numeric_dtype(self.data[col]):
            print(f"\nAggregate for '{col}':")
            print(f"Sum: {self.data[col].sum():.2f}")
            print(f"Mean: {self.data[col].mean():.2f}")
            print(f"Count: {self.data[col].count()}")
        else:
            print(f"Column '{col}' not numeric or not found.")
    
    def statistical_analysis(self):
        print(self.data.columns)
        if self.data.empty:
            print("Load data first.")
            return
    
        col = input("Enter numeric column for stats: ").strip()
        if col in self.data.columns and pd.api.types.is_numeric_dtype(self.data[col]):
            print(f"\nStatistics for '{col}':")
            print(f"Std Dev: {self.data[col].std():.2f}")
            print(f"Variance: {self.data[col].var():.2f}")
            print(f"Quantiles:\n{self.data[col].quantile([0.25, 0.5, 0.75])}")
        else:
            print(f"Column '{col}' not numeric or not found.")

        

def main():
    analyzer = SalesDataAnalyzer()
    while True:
        print("\n Main Menu ")
        print("1. Load Dataset")
        print("2. Explore Data")
        print("3. Perform DataFrame Operations")
        print("4. Clean and Handle Missing Data")
        print("5. Generate Descriptive Statistics")
        print("6. Data Visualization")
        print("7. Save and Display Last Plot")
        print("8. Exit")
        
  
        choice = int(input("Enter your choice: "))

        if choice == 1:
        
            path = input("Enter CSV file path: ")
            analyzer.load_data(path)

        elif choice == 2:
           while True:
                print("\n Explore Data Menu ")
                print("1. Explore Dataset")
                print("2. Combine Data")
                print("3. Remove Duplicate Rows")
                print("4. Back to Main Menu")
            
                sub_choice = int(input("Enter your option: "))
            
                if sub_choice == 1:
                    analyzer.explore_data()
            
                elif sub_choice == 2:
                    analyzer.combine_data()
            
                elif sub_choice == 3:
                    analyzer.remove_duplicates()
            
                elif sub_choice == 4:
                    break
            
                else:
                    print("Invalid option.")



        elif choice == 3:
            while True:
                print("\n DataFrame Operations Menu ")
                print("1. Convert DataFrame to NumPy Array")
                print("2. Search / Sort / Filter Data")
                print("3. Aggregate Functions")
                print("4. Statistical Analysis")
                print("5. Back to Main Menu")
        
                sub_choice = int(input("Enter your option: "))
        
                if sub_choice == 1:
                    analyzer.convert_numpy()
        
                elif sub_choice == 2:
                    analyzer.search_sort_filter()
        
                elif sub_choice == 3:
                    analyzer.aggregate_functions()
        
                elif sub_choice == 4:
                    analyzer.statistical_analysis()
        
                elif sub_choice == 5:
                    break 
        
                else:
                    print("Invalid option! Try again.")


        elif choice == 4:
            analyzer.handle_missing_data()

        elif choice == 5:
            while True:
                print("\n Descriptive Statistics Menu ")
                print("1. Generate Descriptive Statistics")
                print("2. GroupBy and Transform")
                print("3. Back to Main Menu")
        
                sub_choice = int(input("Enter your option: "))
        
                if sub_choice == 1:
                    analyzer.generate_descriptive_stats()
        
                elif sub_choice == 2:
                    analyzer.groupby_and_transform()
        
                elif sub_choice == 3:
                    break 
        
                else:
                    print("Invalid option! Try again.")


        elif choice == 6:
            analyzer.visualize_data()

        elif choice == 7:
            analyzer.save_visualization()

        elif choice == 8:
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```
